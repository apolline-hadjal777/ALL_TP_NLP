## Exercise 1: Function Plotting and Analytical Determination of Critical Points

Determine _analytically_ the coordinates and nature of the critical points of the following functions:

- $$f_1(x,y) = x^3 + y^3 -3x -12y$$
- $$f_2(x,y) = x^3 -12xy + 8y^3$$
- $$f_3(x,y) = {(x-1)}^2 + 10 {(x^2-y)}^2$$

**_Vocabulary:_** An "analytical" determination means that the calculation is done "by hand", without using numerical tools or formal computation.

```python
# Import numpy and pyplot libraries

import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# define the given functions
def f1(x,y) :
    return x**3 + y**3 - 3*x - 12*y
def f2(x,y) :
    return x**3 - 12*x*y + 8*(y**3)
def f3(x,y) :
    return (x-1)**2 + 10*(x**2 - y)**2
```

### Exercise 1 — Analytical Study of Critical Points

We study the following functions:

$f_1(x,y) = x^3 + y^3 - 3x - 12y$

$f_2(x,y) = x^3 - 12xy + 8y^3$

$f_3(x,y) = (x - 1)^2 + 10(x^2 - y)^2$

---

### 1. Function $f_1(x,y) = x^3 + y^3 - 3x - 12y$

#### Gradient

$\partial f_1 / \partial x = 3x^2 - 3 = 3(x^2 - 1)$

$\partial f_1 / \partial y = 3y^2 - 12 = 3(y^2 - 4)$

So the gradient is  

$\nabla f_1(x,y) = \begin{pmatrix} 3(x^2 - 1) \\ 3(y^2 - 4) \end{pmatrix}$

#### Critical points

Set the gradient to zero:

$3(x^2 - 1)=0 \Rightarrow x = \pm 1$

$3(y^2 - 4)=0 \Rightarrow y = \pm 2$

Critical points:

$(-1,-2)$, $(-1,2)$, $(1,-2)$, $(1,2)$

#### Hessian

Second derivatives:

$\partial^2 f_1 / \partial x^2 = 6x$

$\partial^2 f_1 / \partial y^2 = 6y$

$\partial^2 f_1 / \partial x \partial y = 0$

Thus  

$H_{f_1}(x,y) = \begin{pmatrix} 6x & 0 \\ 0 & 6y \end{pmatrix}$

#### Nature of the critical points

- At $(-1,-2)$: both $6x < 0$ and $6y < 0$ → local maximum  
- At $(1,2)$: both $6x > 0$ and $6y > 0$ → local minimum  
- At $(1,-2)$ and $(-1,2)$: one positive, one negative eigenvalue → saddle points

---

### 2. Function $f_2(x,y) = x^3 - 12xy + 8y^3$

#### Gradient

$\partial f_2 / \partial x = 3x^2 - 12y$

$\partial f_2 / \partial y = -12x + 24y^2$

#### Critical points

Solve:

$x^2 = 4y$

$x = 2y^2$

Substitute:

$(2y^2)^2 = 4y \Rightarrow 4y^4 = 4y \Rightarrow y(y^3 - 1)=0$

Thus:  

$y=0 \Rightarrow x=0$  
$y=1 \Rightarrow x=2$

Critical points: $(0,0)$ and $(2,1)$

#### Hessian

$H_{f_2}(x,y) = \begin{pmatrix} 6x & -12 \\ -12 & 48y \end{pmatrix}$

#### Nature

- At $(0,0)$: determinant $= -144 < 0$ → saddle  
- At $(2,1)$: determinant $= 432 > 0$ and trace $=60 > 0$ → local minimum

---

### 3. Function $f_3(x,y) = (x - 1)^2 + 10(x^2 - y)^2$

#### Gradient

$\partial f_3 / \partial x = 2(x - 1) + 40x(x^2 - y)$

$\partial f_3 / \partial y = 20(y - x^2)$

Set $\partial f_3 / \partial y = 0$:

$y = x^2$

Insert into the first equation:

$2(x - 1)=0 \Rightarrow x=1$

Then $y=1$

Critical point: $(1,1)$

#### Hessian

$H_{f_3}(x,y) = \begin{pmatrix} 120x^2 - 40y + 2 & -40x \\ -40x & 20 \end{pmatrix}$

At $(1,1)$:

$H = \begin{pmatrix} 82 & -40 \\ -40 & 20 \end{pmatrix}$

Determinant $=40>0$, trace $=102>0$ → both eigenvalues positive → strict local minimum

Also  

$f_3(x,y) \ge 0$ with equality only at $(1,1)$  

→ global minimum

---

### Final Summary

- For $f_1$:  
  - local max at $(-1,-2)$  
  - local min at $(1,2)$  
  - saddle at $(1,-2)$ and $(-1,2)$  

- For $f_2$:  
  - saddle at $(0,0)$  
  - local min at $(2,1)$  

- For $f_3$:  
  - global min at $(1,1)$

```python
def df1(x, y):
    # Partial derivative with respect to x
    df_dx = 3*(x**2 - 1)

    # Partial derivative with respect to y
    df_dy = 3*(y**2 - 4)

    return [df_dx, df_dy]
```

**3.4** The instruction block below provides the template for the function _d2f1_, which calculates and returns the Hessian matrix of $f_1$. Complete this instruction block so that _d2f1_ does what is expected. Then use this function to verify the nature of the critical points of $f_1$.

```python
def d2f1(x, y):
    # Second derivative wrt x
    d2f_dxdx = 6*x

    # Mixed derivative
    d2f_dxdy = 0
    d2f_dydx = 0

    # Second derivative wrt y
    d2f_dydy = 6*y

    return [[d2f_dxdx, d2f_dxdy],
            [d2f_dydx, d2f_dydy]]
```

```python
# Template for calculating eigenvalues of a matrix, to be adapted:
matrice = [[1,2],[3,4]]
valProp = np.linalg.eigvals(matrice)
print('Eigenvalues of the matrix {} :'.format(matrice))
print('  vp1 = {}'.format(valProp[0]))
print('  vp2 = {}'.format(valProp[1]))
```

```python
# Determine the nature of the four critical points of f1

crit_points = [(-1, -2), (-1, 2), (1, -2), (1, 2)]

for (x0, y0) in crit_points:
    print(f"Critical point: ({x0}, {y0})")

    H = np.array(d2f1(x0, y0))
    valProp, vectProp = np.linalg.eig(H)

    print("  vp1 =", valProp[0])
    print("  vp2 =", valProp[1])

    if valProp[0] > 0 and valProp[1] > 0:
        print("  → local minimum\n")
    elif valProp[0] < 0 and valProp[1] < 0:
        print("  → local maximum\n")
    else:
        print("  → saddle point\n")

# Calculate the Hessian matrix
matrice1 = d2f1 (-1,-2)
# Calculate the eigenvalues of the Hessian matrix, then test their sign
valProp = np.linalg.eigvals(matrice1)
print('Eigenvalues of the matrix {} :'.format(matrice1))
print('  vp1 = {}'.format(valProp[0]))
print('  vp2 = {}'.format(valProp[1]))
```
